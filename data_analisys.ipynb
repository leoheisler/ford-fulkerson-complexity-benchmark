```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Carregar o CSV
csv_path = './results/heat_map.csv'  # Altere este caminho conforme necessário
data = pd.read_csv(csv_path)

# Criar uma tabela pivot com as variáveis 'A' e 'B' para construir o heatmap
heatmap_data = data.pivot(index="B", columns="A", values="PROB")

# Criar o gráfico de calor
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm_r", fmt=".3f", cbar_kws={'label': 'Probability'})

# Definir títulos e rótulos
plt.title("Heatmap of Probability by A and B", fontsize=16)
plt.xlabel("A", fontsize=12)
plt.ylabel("B", fontsize=12)
plt.gca().invert_yaxis()

# Exibir o gráfico
plt.show()
```
