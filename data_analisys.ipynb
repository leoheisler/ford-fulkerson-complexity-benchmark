```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carrega o CSV
df = pd.read_csv("results/fattest/fattest_linegraph.csv")

# Calcula a complexidade teórica
df["complexity"] = df["VERTEX_NUM"] * df["EDGES_NUM"] * (df["VERTEX_NUM"] + df["EDGES_NUM"])

# Ordena pelo campo de complexidade
df_sorted = df.sort_values("complexity")

# Plota os dados ordenados
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["complexity"], df_sorted["DURATION_COMPLEXITY_RATIO"], marker='o', linestyle='-')
plt.xlabel("Complexidade (V * E * (V + E))")
plt.ylabel("DURATION_COMPLEXITY_RATIO")
plt.title("Line Graphs")
plt.grid(True)
plt.tight_layout()
plt.show()

df = pd.read_csv("results/fattest/fattest_meshgraphs.csv")

# Calcula a complexidade teórica
df["complexity"] = df["VERTEX_NUM"] * df["EDGES_NUM"] * (df["VERTEX_NUM"] + df["EDGES_NUM"])

# Ordena pelo campo de complexidade
df_sorted = df.sort_values("complexity")

# Plota os dados ordenados
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["complexity"], df_sorted["DURATION_COMPLEXITY_RATIO"], marker='o', linestyle='-')
plt.xlabel("Complexidade (V * E * (V + E))")
plt.ylabel("DURATION_COMPLEXITY_RATIO")
plt.title("Mesh Graphs")
plt.grid(True)
plt.tight_layout()
plt.show()

df = pd.read_csv("results/fattest/fattest_rand2graphs.csv")

# Calcula a complexidade teórica
df["complexity"] = df["VERTEX_NUM"] * df["EDGES_NUM"] * (df["VERTEX_NUM"] + df["EDGES_NUM"])

# Ordena pelo campo de complexidade
df_sorted = df.sort_values("complexity")

# Plota os dados ordenados
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["complexity"], df_sorted["DURATION_COMPLEXITY_RATIO"], marker='o', linestyle='-')
plt.xlabel("Complexidade (V * E * (V + E))")
plt.ylabel("DURATION_COMPLEXITY_RATIO")
plt.title("Rand2 Graphs")
plt.grid(True)
plt.tight_layout()
plt.show()
```
